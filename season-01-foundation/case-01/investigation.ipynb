### Sección #2 - Reconocimiento de Evidencia

```python
# Información general del dataset
df_transacciones.describe()
```

```python
# Comprobación de nulos
print(df_transacciones.isnull().sum())
```

```python
# Valores únicos de la columna Class

df_transacciones['Class'].unique()
```

```python
#  Determinar el rango de montos (mínimo, máximo, promedio)

print(f"Monto Mínimo: {df_transacciones['Amount'].min()}")
print(f"Monto Máximo: {df_transacciones['Amount'].max()}")
print(f"Monto Promedio: {df_transacciones['Amount'].mean()}")
```

### Sección #3 - Siguiendo la Pista de Pemberton

```python
# Filtrar solo transacciones fraudulentas

df_transacciones_fraude = df_transacciones[df_transacciones['Class'] == 1]
df_transacciones_normales = df_transacciones[df_transacciones['Class'] == 0]

df_transacciones_fraude.head()
```

```python
# Transacciones fraudulentes vs normales - Suma total
df_transacciones['Class'].value_counts()
```

```python
# Porcentaje de transacciones fraudulentas

total_transacciones = len(df_transacciones)
total_transacciones_fraude = len(df_transacciones_fraude)

porcentaje_transacciones_fraude = (total_transacciones_fraude / total_transacciones) * 100

print(f'Porcentaje de transacciones fraudulentas: {porcentaje_transacciones_fraude:.2f}%')
```

```python
# Promedio de transacciones para cada clase

df_transacciones.groupby('Class')['Amount'].mean()
```

### Sección #4 - Análisis Preliminar

```python
# Top 5 de transacciones fraudulentas

df_transacciones_fraude['Amount'].sort_values(ascending=False).head()
```

```python
# Distribución de montos según la clase de la transacción

df_transacciones.boxplot(column='Amount', by='Class')
```
